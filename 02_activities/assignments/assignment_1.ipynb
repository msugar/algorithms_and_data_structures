### Define the Problem
- Paraphrase the problem in your own words


*Your answer here*

We're tasked with finding every possible path you can take from the root of a binary tree to any leaf node. In other words:
1. Start at the top (root) node
2. Keep walking down the tree, choosing either left or right child at each step
3. Stop when you hit a dead end (leaf node - a node with no children)
4. Record the sequence of node values you encountered along each complete path

For example, if we have this tree:
```
     1
    / \
   2   3
  /\   /\
 4  5 6  7
```
The answer would be all these paths:
- `[1, 2, 4]` by going root->left->left
- `[1, 2, 5]` by going root->left->right
- `[1, 3, 6]` by going root->right->left
- `[1, 3, 7]` by going root->right->right

Each path must:
- Start at the root
- End at a leaf
- Include every node value encountered along the way
- Follow valid parent-child connections

It's essentially asking us to document every possible way to walk from the top of the tree to any bottom node, recording the numbers we see along each walk.

````python
# In a real interview, the code in this cell could be given or assumed. 
# I'm conding it here to make it possible to test the code with unit tests later.

from typing import List, Optional
from typing_extensions import Self

# Definition for a binary tree node.
class TreeNode:
    """
    A binary tree node with integer value and optional left and right children.
    
    Attributes:
        val (int): The node's value
        left (TreeNode | None): Left child node
        right (TreeNode | None): Right child node
    """    
    def __init__(self, val: int = 0, left: Optional['TreeNode'] = None, right: Optional['TreeNode'] = None) -> Self:
        self.val = val
        self.left = left
        self.right = right
        
    def is_leaf(self) -> bool:
        """Returns True if the node has no children."""
        return self.left is None and self.right is None

    def __repr__(self):
        """String representation of the node."""
        return str(self.val)

    @classmethod
    def build_from_breadth_first_traversal(cls, bfs_traversal: List[Optional[int]]) -> Optional['TreeNode']:
        """
        Builds a binary tree from a level-order traversal array.
        None values in the array represent empty nodes.
        
        Args:
            bfs_traversal: List of integers or None representing the tree in level-order
            
        Returns:
            The root node of the constructed tree, or None if input is empty
            
        Raises:
            ValueError: If the input array is invalid for tree construction
        """ 
        if not bfs_traversal:
            return None
        
        # Validate input
        if not all(isinstance(x, (int, type(None))) for x in bfs_traversal):
            raise ValueError("Array must contain only integers or None values")
        
        nodes = []
        for index, val in enumerate(bfs_traversal):
            if val is not None:
                node = cls(val)
                nodes.append(node)
                if index > 0:
                    parent_index = (index - 1) // 2
                    if index % 2 == 1:
                        nodes[parent_index].left = node
                    else:
                        nodes[parent_index].right = node
            else:
                nodes.append(None)

        return nodes[0] if len(nodes) > 0 else None
````

````python
# Your answer here

def bt_path(root: Optional[TreeNode]) -> List[List[int]]:
    """
    Returns all root-to-leaf paths in a binary tree.
    
    Args:
        root: Root node of the binary tree
        
    Returns:
        List of all paths from root to leaves, where each path is a list of node values
    """
    if not root:
        return []    
  
    def visit(node: TreeNode, current_path: List[int], all_paths: List[List[int]]) -> None:
        """
        DFS helper function to traverse the tree and collect paths.
        
        Args:
            node: Current node being visited
            current_path: Path from root to current node
            all_paths: Collection of all complete root-to-leaf paths
        """        
        current_path.append(node.val)
  
        if node.is_leaf():
            all_paths.append(current_path.copy())
        else:
            if node.left:
                visit(node.left, current_path, all_paths)
            if node.right:
                visit(node.right, current_path, all_paths)
  
        current_path.pop()
        
    all_paths = []
    visit(root, [], all_paths)
    return all_paths
````

````python
# This is just to impress the interviewer. :-)

import unittest

class TestBinaryTreePathsToLeaves(unittest.TestCase):
    """Test cases for binary tree path finding algorithm."""

    def test_example_1(self):
        """Test Example 1."""
        input = [1, 2, 2, 3, 5, 6, 7]
        root = TreeNode.build_from_breadth_first_traversal(input)
        expected = [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]
        self.assertCountEqual(bt_path(root), expected)
        
    def test_example_2(self):
        """Test Example 2."""
        input = [10, 9, 7, 8]
        root = TreeNode.build_from_breadth_first_traversal(input)
        expected = [[10, 7], [10, 9, 8]]
        self.assertCountEqual(bt_path(root), expected)
        
    def test_empty_tree(self):
        """Test handling of empty tree."""
        root = None
        self.assertEqual(bt_path(root), [])
            
    def test_single_node(self):
        """Test tree with only root node."""
        root = TreeNode(1)
        self.assertEqual(bt_path(root), [[1]])
        
    def test_invalid_input(self):
        """Test invalid input handling."""
        with self.assertRaises(ValueError):
            TreeNode.build_from_breadth_first_traversal(['invalid'])
            
    def test_none_values(self):
        """Test tree construction with None values."""
        input = [1, None, 3]
        root = TreeNode.build_from_breadth_first_traversal(input)
        expected = [[1, 3]]
        self.assertEqual(bt_path(root), expected)
        
    def test_balanced_tree(self):
        """Test balanced tree with multiple paths."""
        input = [1, 2, 3, 4, 5, 6, 7]
        root = TreeNode.build_from_breadth_first_traversal(input)
        expected = [[1, 2, 4], [1, 2, 5], [1, 3, 6], [1, 3, 7]]
        self.assertCountEqual(bt_path(root), expected)
    
    def test_unbalanced_tree(self):
        """Test unbalanced tree."""
        input = [1, 2, None, 3, None, None, None, 4, 5]
        root = TreeNode.build_from_breadth_first_traversal(input)
        expected = [[1, 2, 3, 4], [1, 2, 3, 5]]
        self.assertEqual(bt_path(root), expected)
````

### Solution Explanation
- Explain why your solution works


*Your answer here*

To understand how this code finds all paths from the root to any leaf in a tree, imagine you're at the entrance of a maze with no exits and you want to map all the paths from the entrance to any dead end.

The main function bt_path starts by handling an empty tree - if there's no tree, there are no paths, so it returns an empty list.
The real work happens in the helper function visit. Think of it like exploring with a piece of string - you tie the string at the entrance and unroll it as you walk:
1. When you reach a new room (node), you add its number to your current path (current_path.append).
2. Then you check: If it's a dead end (leaf node), you've found a complete path! Make a copy of your current path and save it. If not, explore any available doorways - first left, then right - taking your string with you.
3. After exploring all possibilities from this room, you step back (backtrack) by removing this room's number from your path (current_path.pop). This is like rewinding your string to try a different route.

The clever part is the backtracking - by removing each room number as we step back, we keep our path accurate for exploring other routes. We're reusing the same path list instead of making new ones for every possible route.
At the end, we have a list of all possible paths from entrance to every dead end in our maze.

Example:
```
     1
    / \
   2   3
  /     \
 4       5
```

The function would:
1. Start at 1, add it to path: `[1]`
1. Go left to 2: `[1,2]`
1. Go left to 4: `[1,2,4]`
1. Found leaf! Save this path
1. Step back: `[1,2]`
1. No right path at 2, step back: `[1]`
1. Go right to 3: `[1,3]`
1. Go right to 5: `[1,3,5]`
1. Found leaf! Save this path
1. Done! Return `[[1,2,4], [1,3,5]]`

### Time and Space Complexity
- Explain the problemâ€™s time and space complexity

*Your answer here*
 
*Time Complexity:*  `O(N)`, where `N` is the number of nodes.
- The algorithm visits each node exactly once
    
*Space Complexity:* `O(P) * O(D)`, where `P` is the number of paths (or leaves) and `D` is the depth of the tree.
- We can't have more than (N+1)/2 leaves in a binary tree. Therefore, the maximum number of elements in `all_paths` is `O(P) = O((N+1)/2) = O(N)`
- The depth (or height) of a *balanced* tree is `O(log(N))`, so for balanced trees `O(D) = O(log(N))` and total space complexity would be `O(N*log(N))`
- But an unbalanced tree can have depth `O(N)`, so the in the worst-case scenario the space complexity is `O(N) * O(N) = O(N^2)`
    

### Alternative Solution Proposal
-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


*Your answer here*

Here is an alternative approach using a generator, which is more memory-friendly. Think of it like writing directions for exploring a building, but instead of writing down all paths at once, you guide someone through one complete path at a time.

Here's the thinking:
1. Instead of collecting all paths in a list, we'll "yield" each path as we find it
2. Think of "yield" like saying "here's one path I found" before continuing to look for more
3. The person following our directions can deal with each path as it's found, rather than waiting for all of them

The big difference is:
- Original solution: Like taking photos of every possible path and giving someone the whole photo album at the end
- Generator solution: Like walking with someone and pointing out each complete path as you discover it

The advantages:
- Uses less memory because you only keep track of the current path you're exploring
- Good for huge trees where storing all paths at once would be too much
- The person using our code can start processing paths immediately, without waiting for all of them

The trade-off:
- You can only look at each path once as it's generated (unless you save them yourself)
- Can't know how many paths there are until you've seen them all

It's like the difference between:
- Writing down all the ways to get through a maze, then giving someone the complete list
- Walking through the maze with someone and telling them "here's one way!" each time you find an exit, then backtracking to find more

For our classmates: When coding this up, you'll use Python's `yield` keyword instead of building a list. Each time you find a leaf node, instead of appending to a list, you'll `yield` that path. The rest of the logic stays pretty similar!
